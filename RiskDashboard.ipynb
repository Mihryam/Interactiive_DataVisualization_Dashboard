```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load your credit risk data (replace with your dataset)
df = pd.read_csv('UCI_Credit_Card.csv')


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a function to update the dashboard
def update_dashboard():
    # Clear the previous plots
    plt.clf()

    # Update default font settings to make text bolder
    plt.rcParams.update({'font.weight': 'bold', 'font.size': 14})

    # Create subplots for distribution, credit risk distribution, credit risk by education, credit risk by marriage, and correlation heatmap
    plt.figure(figsize=(30, 10))  # Increase the figsize for larger charts

    # Distribution plot
    plt.subplot(151)
    sns.histplot(df["AGE"], kde=True)
    plt.title("Distribution of AGE")

    # Visualize credit risk distribution
    plt.subplot(152)
    sns.countplot(x='default.payment.next.month', data=df)
    plt.title('Credit Risk Distribution')

    # Visualize credit risk by education level
    plt.subplot(153)
    sns.countplot(x='EDUCATION', hue='default.payment.next.month', data=df)
    plt.title('Credit Risk by Education Level')

    # Visualize credit risk by marriage status
    plt.subplot(154)
    sns.countplot(x='MARRIAGE', hue='default.payment.next.month', data=df)
    plt.title('Credit Risk by Marriage Status')

    # Adjust layout
    plt.tight_layout()

    # Display the updated dashboard
    plt.show()

# Initial dashboard
update_dashboard()

# Interactive loop for user input
while True:
    print("Options:")
    print("1. Update the dashboard")
    print("2. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        update_dashboard()
    elif choice == "2":
        break
    else:
        print("Invalid choice. Please enter 1 to update the dashboard or 2 to exit.")

print("Dashboard closed.")
```
